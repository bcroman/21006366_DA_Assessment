**Add Libraries**

```r
install.packages('corrplot')
install.packages('PerformanceAnalytics')
library('corrplot')
library('PerformanceAnalytics')
library('lattice')
```

**Read in CSV File**

```r
raw_data <- read.csv('https://raw.githubusercontent.com/bcroman/21006366_DA_Assessment/refs/heads/main/views/Crime_Weather_2014_2016.csv')
```

**Print Data**

```r
summary(raw_data)
head(raw_data)
```

**Clean Up Dataset**

```r
keep_columns <- c('year', 'month', 'total_crimes', 'major_category', 'avg_temp', 'avg_dewp', 'avg_visib', 'avg_wdsp', 'avg_mxpsd', 'avg_gust', 'avg_max_temp', 'avg_min_temp', 'avg_prcp', 'fog_days')
compressed_data <- raw_data[keep_columns]
summary(compressed_data)
```

**Plotting Data**
- Crime Vs Weather

```r
plot(compressed_data$total_crimes, compressed_data$avg_temp, main = "Total Crimes Vs Avg Temp")

linear_prediction <- lm(total_crimes ~ avg_temp, data = compressed_data)
abline(linear_prediction)

print(linear_prediction)
```

```r
plot(compressed_data$total_crimes, compressed_data$avg_prcp, main = "Total Crimes Vs Avg Rainfall")

linear_prediction <- lm(total_crimes ~ avg_prcp, data = compressed_data)
abline(linear_prediction)

print(linear_prediction)
```

```r
plot(compressed_data$total_crimes, compressed_data$fog_days, main = "Total Crimes Vs Fog Days")

linear_prediction <- lm(total_crimes ~ fog_days, data = compressed_data)
abline(linear_prediction)

print(linear_prediction)
```

Plot Charts of Total Crime Over Time

```r
plot(compressed_data$total_crimes, compressed_data$year, main = "Total Crimes Vs Year")
```

```r
plot(compressed_data$total_crimes, compressed_data$month, main = "Total Crimes Vs Month")
```

**Correlation Matrix**
- Steamline Matrix
- Weather Matrix Value
  

```r
keep_columns <- c('year', 'month', 'total_crimes', 'avg_temp', 'avg_dewp', 'avg_visib', 'avg_wdsp', 'avg_mxpsd', 'avg_gust', 'avg_max_temp', 'avg_min_temp', 'avg_prcp', 'fog_days')
steamline_data <- raw_data[keep_columns]
summary(steamline_data)
```

```r
core_data <- cor(steamline_data)
```
