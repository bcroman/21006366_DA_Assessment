**Read in CSV File**

```r
raw_data <- read.csv('https://raw.githubusercontent.com/bcroman/21006366_DA_Assessment/refs/heads/main/views/Crime_Weather_2014_2016.csv')
```

**Print Data**

```r
summary(raw_data)
head(raw_data)
```

**Clean Up Dataset**

```r
keep_columns <- c('year', 'month', 'total_crimes', 'avg_temp', 'avg_dewp', 'avg_visib', 'avg_wdsp', 'avg_mxpsd', 'avg_gust', 'avg_max_temp', 'avg_min_temp', 'avg_prcp', 'fog_days')
compressed_data <- raw_data[keep_columns]
summary(compressed_data)
```

**Plotting Data**
- Crime Vs Weather

```r
plot(compressed_data$total_crimes, compressed_data$avg_temp, main = "Total Crimes Vs Avg Temp")
```

```r
plot(compressed_data$total_crimes, compressed_data$avg_prcp, main = "Total Crimes Vs Avg Rainfall")
```

```r
plot(compressed_data$total_crimes, compressed_data$fog_days, main = "Total Crimes Vs Fog Days")
```

**Correlation Matrix**
- Weather Matrix Value

```r
crime_temp <- cor(compressed_data$total_crimes, compressed_data$avg_temp)
print(paste('Total Crimes Vs Avg Temp:', crime_temp))
```

```r
crime_rain <- cor(compressed_data$total_crimes, compressed_data$avg_prcp)
print(paste('Total Crimes Vs Avg Rainfall:', crime_rain))
```

```r
crime_fog <- cor(compressed_data$total_crimes, compressed_data$fog_days)
print(paste('Total Crimes Vs Fog Days:', crime_fog))
```
