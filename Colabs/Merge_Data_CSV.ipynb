Import Libarys

```python
import pandas as pd
```

Get Weather and Crime CSV Files and Store to Varbailes

```python
crime_url = "https://raw.githubusercontent.com/bcroman/21006366_DA_Assessment/refs/heads/main/views/London/London%20Data%202014%20-%202016.csv"
weather_url = "https://raw.githubusercontent.com/bcroman/21006366_DA_Assessment/refs/heads/main/views/Weather/Weather%20Monthly%20Data.csv"

crime = pd.read_csv(crime_url)
weather = pd.read_csv(weather_url)
```

Merge Weather and Crime base on Year and Month

```python
merged = pd.merge(crime, weather, on=["year", "month"], how="left")

merged.head()
```

Save merge table as a CSV file

```python
merged.to_csv("merged_data.csv", index=False)
```
